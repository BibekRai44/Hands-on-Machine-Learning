```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df=pd.read_csv("ds_salaries.csv")
```

```python
df
```

```python
df.dtypes
```

```python
plt.plot(df['work_year'],df['salary_in_usd'],c='b',linestyle='dashed',marker='*',markersize='5')
plt.xlabel('Work Year')
plt.ylabel("salary in usd")
plt.title("Work year vs Salary",c='g')
```

```python

x = np.arange(1,8) 
y = 3 * x + 5 
plt.title("Matplotlib demo") 
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(x,y) 
plt.show()
```

```python
x=np.arange(1,2*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sample Example")
```

```python
plt.bar(df['work_year'],df['salary_in_usd'],color='black',align='center')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sample Example")
```
